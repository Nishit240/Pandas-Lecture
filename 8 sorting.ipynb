* sorting data
* SORTING DATA 1 COLUMN sort_values()
* df.sort_values(by="Column Name", ascending = True/False, inplace = True)

```python
import pandas as pd
```

```python
data = {
    "Name": ['Arun', 'Varun', 'Karun'],
    "Age": [23,54,12],
    "Salary": [10000, 20000,30000]
    }

df = pd.DataFrame(data)
```

```python
df.sort_values(by='Age', ascending=False, inplace=False )
```

```python
df.sort_values(by='Age', ascending=True, inplace=False )
```

```python
df.sort_values(by=['Age', 'Salary'], ascending=False, inplace=False )
```
