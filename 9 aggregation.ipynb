* df ["Column Name"].mean()
* df ["Column Name"].sum()
* df ["Column Name"].min()
* df ["Column Name"].max()

```python
import pandas as pd
```

```python
data = {
"Name": ['Arun', 'Varun', 'Karun'],
"Age": [28,34,22], 
"Salary": [10000, 20000,30000]
}
df = pd.DataFrame(data)
```

```python
avg_salary = df['Salary'].mean()
print(avg_salary)
```

```python
sum_salary = df['Salary'].sum()
print(sum_salary)
```

```python
max_salary = df['Salary'].max() 
print(max_salary)
```

```python
min_salary = df['Salary'].min()
print(min_salary)
```
