* NaN (Not a Number)
* None (for object data type)

* isnull()
* true - NaN is present
* False - value is present


```python
import pandas as pd
```

```python
data = {
    "Name": ['Ram', None, 'Ghanshyam', 'Dhanshyam', 'Aditi', 'Jagdish', 'Raj', 'Simran'],
    "Age": [28,34,None,30,29,40,25,None],
    "Salary": [50000,60000,45000,52000,49000, 70000,48000,None],
    "Performance Score": [None,90,78,92,88,95,80,89]
}

df = pd.DataFrame(data)
```

```python
print(df.isnull(), "\n")
print(df.isnull().sum())
```

* drope the whole row from the table by using dropna()  

* axis = 0 (drop the rows where Null is present)
* axis = 1 (drop the coloum where Null is present)

```python
df.dropna(axis=0, inplace=False)
```

```python
df.dropna(axis=1, inplace=False)
```

* replace the value with any other thing like mean 0 max  
* fillna()


```python
df.fillna(0, inplace=False)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=False) 

df['Salary'].fillna(df['Salary'].mean(), inplace=False) 
```

### Interpolation
It is the mathematical process of estimating unknown values that fall between known data points

* interpolate() 
#### Key Methods
* linear
* time
* polynomial
* nearest

#### Parameters
* axis = fill missing values along columns (axis=0) or across rows (axis=1)
* limit = Sets a maximum number of consecutive NaNs that can be filled in one go
* limit_direction = Determines the direction of the filling ('forward', 'backward', or 'both')
* inplace = True or False


```python
data = {
    "Name": ['Ram', 'Palak', 'Ghanshyam', 'Dhanshyam', 'Aditi', 'Jagdish', 'Raj', 'Simran'],
    "Age": [28,34,None,30,29,40,25,None],
    "Salary": [50000,60000,45000,52000,49000, 70000,48000,None],
    "Performance Score": [None,90,78,92,88,95,80,89]
}

df = pd.DataFrame(data)
```

```python
df.interpolate(axis=0, inplace=False)
```

#### Linear

```python
data = {
"Time": [1,2,3,4,5],
"Value": [10, None, 30, None, 50]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['Value'] = df['Value'].interpolate(method='linear')
print(df)
```
