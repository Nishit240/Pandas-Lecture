# Introduction to Pandas
## What is Pandas?
A powerful library for data manipulation and analysis in Python.
Built on top of NumPy, provides data structures like Series and DataFrame.
## Why use Pandas?
* Easy data cleaning and transformation.
* Efficient handling of large datasets.
* Integration with other Python libraries like Matplotlib and Seaborn for visualization.

```python
# !pip install pandas
import pandas as pd
```

```python

data = {
    'Name': ['Nishit', 'Palak', 'Rohan'],
    'Age': [21, 22, 29],
    'Department': ['IT', 'HR', 'Finance']
}
print(type(data))
df = pd.DataFrame(data)
print(type(df))
print(df)

print(df['Name'])

# Employees with Age > 21 
print(df[df['Age'] > 21])

df['Salary'] = [50000, 60000, 55000]
print(df)

# Increase salary by 10%
df['New_Salary'] = df['Salary'].apply(lambda x: x * 1.1)
print(df)

# Average salary by Department
grouped = df.groupby('Department')['Salary'].mean()
print(grouped)


data1 = {
    'Name': ['Nishit', 'Palak', 'Rohan', 'Sneha', 'Amit'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [50000, 60000, 55000, 65000, 62000]
}

df = pd.DataFrame(data1)
print(df['Department'].value_counts())

avg_salary = df.groupby('Department')['Salary'].mean()
print(print("Average Salary per Department:\n", avg_salary)
)

# Group by Department and find max salary
max_salary_per_dept = df.groupby('Department')['Salary'].transform('max')

# Filter employees with salary equal to max salary in their department
top_earners = df[df['Salary'] == max_salary_per_dept]
print("\nTop Earners per Department:\n", top_earners)

```
